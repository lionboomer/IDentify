```python
import pymongo
import base64
import numpy as np
from PIL import Image
from io import BytesIO
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# MongoDB-Verbindung herstellen
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["fingerprintDB"]
collection = db["fingerprints"]
```

```python
# Daten aus der Datenbank lesen
users_data = collection.find()
print("Datenbankabfrage erfolgreich!")
#Daten anzeigen die geladen wurden 
for user in users_data:
    print(user)
```

```python
# berprfen, ob es mindestens zwei Benutzer mit 20 Canvas-Eintrgen gibt
count = 0

# Ensure that users_data is defined and contains the necessary data
users_data = collection.find()

for user in users_data:
    print(user["username"], len(user["canvases"]))
    if len(user["canvases"]) == 19:
        count += 20
    if count >= 2:
        break

if count >= 2:
    print("Es gibt mindestens zwei Benutzer mit 20 Canvas-Eintrgen.")
else:
    print("Es gibt weniger als zwei Benutzer mit 20 Canvas-Eintrgen.")
```
